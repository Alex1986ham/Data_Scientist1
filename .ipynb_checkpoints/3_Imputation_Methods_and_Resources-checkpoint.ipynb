```python
# One solution for nan Values is to replace these with imputing them with mean, median or mode of a columns intou the missing values.
```

```python
import pandas as pd
import numpy as np
#import ImputationMethods as t

df = pd.DataFrame({'A':[np.nan, 2, np.nan, 0, 7, 10, 15],
                   'B':[3, 4, 5, 1, 2, 3, 5],
                   'C':[np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan],
                   'D':[np.nan, True, np.nan, False, True, False, np.nan],
                   'E':['Yes', 'No', 'Maybe', np.nan, np.nan, 'Yes', np.nan]})

df
```

```python
new_df = df.drop('C', axis = 1)
new_df.head()
```

```python
# TRY TO REPLACE NAN WITH MEAN
fill_mean = lambda col: col.fillna(col.mean())
#fill_mean.head()

try:
    new_df[['A', 'B']].apply(fill_mean, axis=0)
except:
    print('That broke')
    
new_df2 = new_df[['A', 'B', 'D']].apply(fill_mean, axis=0)
```

```python
new_df2.head()
```

```python
# TRY TO REPLACE NAN WITH MODE
fill_mode = lambda col: col.fillna(col.mode()[0])

new_df.apply(fill_mode, axis=0)

# See that mode works for the boolean better than mean
```
